```python
import plotly.express as px
import pandas as pd

df = pd.read_csv('/content/LRP_20k_map.csv')
df.dropna(axis=0, inplace=True ) 

#print(df)
```

```python
import plotly.express as px

fig = px.scatter(df, x="shannon_normalized_alpha", y="PD_whole_tree_normalized_alpha", color="Status")
fig.show()
```

```python
import plotly.express as px
fig = px.scatter(df, x="shannon_normalized_alpha", y="PD_whole_tree_normalized_alpha", color="Status", size='Cd_tissue', hover_data=['Site'])
fig.show()
```

```python
import plotly.express as px
fig = px.scatter(df, x="shannon_normalized_alpha", y="PD_whole_tree_normalized_alpha", color="Status", marginal_y="violin",
           marginal_x="box", trendline="ols", template="simple_white")
fig.show()
```

```python
import plotly.express as px
fig = px.histogram(df, x="Site", color="Status", barmode="group")
fig.show()
```

```python
import plotly.express as px
fig = px.histogram(df, x="shannon_normalized_alpha", color="Status")
fig.show()
```

```python
import plotly.express as px
fig = px.sunburst(df, path=['Site', 'Status', 'Lineage'], color='Status')
fig.show()
```

```python
fig = px.scatter_3d(df, x="shannon_normalized_alpha", y="pH_soil", z='PD_whole_tree_normalized_alpha',
              color='Site', size_max=18,
              symbol='Status', opacity=0.7)

# tight layout
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
```

```python
df_poll = df[df['Status'] == 'Polluted']

fig = px.scatter_matrix(df_poll,
    dimensions=["Al_tissue", "Fe_tissue", "Cd_tissue", "As_tissue", "pH_soil"],
    color="Status" )
fig.show()
```

```python
df = px.data.iris()
fig = px.parallel_coordinates(df, color="species_id", labels={"species_id": "Species",
                "sepal_width": "Sepal Width", "sepal_length": "Sepal Length",
                "petal_width": "Petal Width", "petal_length": "Petal Length", },
                             color_continuous_scale=px.colors.diverging.Tealrose,
                             color_continuous_midpoint=2)
fig.show()
```

```python
df = px.data.wind()
fig = px.scatter_polar(df, r="frequency", theta="direction",
                       color="strength", symbol="strength", size="frequency",
                       color_discrete_sequence=px.colors.sequential.Plasma_r)
fig.show()
```
