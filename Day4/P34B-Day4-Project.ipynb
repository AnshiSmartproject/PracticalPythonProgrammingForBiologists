# Day 4 Project - Visualisation of lumbricid data

We have a range of measured data from a study on Earthworm gut microbiota and associated anthropogenic contaminants both from tissue and soil measurements. 

You have been provided with two files:
- ```Day4-Project-LRP_metadata``` - A large wide ranging dataset for each individual including diversity calculations, soil and tissue metal levels, pH etc.
- ```Day4-Project-LRP_phyla``` - Top level bacterial phyla measurements for each individual

The objective is to load the data and identify key comparisons to represent visually using matplotlib/seaborn/plotly. Note that this is real data and therefore is not perfectly clean.

Some potential questions. Consider which plots you would use to represent it. Test a few!
- Which sites have the highest/lowest diversity?
- Do diversity calculations correspond with specific metals?
- How closely does tissue metal load correspond with soil content
- Do phylogenetic groups associate with particular sites?

In microbial community analysis, it is a classic aproach to create stacked bar-plots of taxa per site, so that is is also a good task to attempt!
